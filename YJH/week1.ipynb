```python
import numpy as np
from functools import reduce

### 이 블럭에 각 함수를 정의합니다. ###

"""
시그모이드 함수를 정의하는 예시입니다.

def sigmoid(x) :
    
    sigmoid = 1 / (1+np.exp(-x))
    
    return sigmoid
    
"""
tanh =  lambda x :2/(1+np.exp(-2*x)) - 1

softsign =  lambda x : x / (1 + np.abs(x))

def relu(x):
    return list(map(lambda a: 0 if a < 0 else a, x))

def selu(x, alpha = 1.67326):
    return list(map(lambda a: 1.0507*alpha*(np.exp(a) - 1) if a < 0 else 1.0507*a, x))

gaussian = lambda x : np.exp(-(x**2))

```

```python
import matplotlib.pyplot as plt
% matplotlib inline

### 이 블럭에 함수들을 전부 한 그림에 나타나도록 그려봅시다. ###
x = np.arange(-3, 3, 0.1)

plt.plot(x, tanh(x), 'r', label='tanh')
plt.plot(x, softsign(x), 'g', label='softsign')
plt.plot(x, relu(x), 'b', label='relu')
plt.plot(x, selu(x), 'c', label='selu')
plt.plot(x, gaussian(x), 'm', label='gaussian')

plt.axvline(linewidth=1, color='k')
plt.axhline(linewidth=1, color='k')

fig = plt.gcf() #현재 figure에 불러오기
fig.set_size_inches(18.5, 10.5) #크기 바꾸기(inch 단위)
fig.savefig('week1.png', dpi=100) #현재 figure 저장하기

plt.legend()
plt.show()
```
