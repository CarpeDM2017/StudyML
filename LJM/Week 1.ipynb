```python
# 함수를 정리해 보아요!
import numpy as np

def tanh(x):
    return 2/(1+np.exp(-2*x))-1

def softsign(x):
    return x/ (1+np.abs(x))

def relu(x):
    # if x>=0,relu(x)=x  and if x<0, relu(x)=0
    return (x>=0)*x

def gaussian(x): 
    return np.exp(-np.square(x))

def selu(a,x):
    lambd=1.0507
    a=1.67326
    # if x>=0, selu(a,x)=lambd * x and if x<0, lambd * a * (exp(x)-1)
    return lambd * ((x>=0) * x + (x<0) * a * (np.exp(x)-1))

```

```python
# 그래프로 그려 보아요!
import matplotlib.pyplot as p
% matplotlib inline

x=np.linspace(-5,5,100) 

p.plot(x, tanh(x), x, softsign(x), x, relu(x), x, gaussian(x), x, selu(a,x))
p.show()
```
